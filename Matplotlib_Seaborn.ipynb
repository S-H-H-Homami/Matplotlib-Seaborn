```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
years=[1960, 1970, 1980, 1990, 2000, 2010, 2016 ]

pop=[21.91, 28.51, 38.67, 56.23, 66.13, 74.57, 80.28]
```

```python
plt.plot(years, pop)
```

```python
plt.scatter(years, pop)
```

```python
plt.hist(pop, bins=3) #best value for bins=sqrt(number of datas)
```

```python
plt.hist(pop, bins='auto') 
```

```python
plt.pie(pop, labels=years)
```

```python
plt.figure(figsize=(8,4), dpi=100, facecolor='cyan', edgecolor='black')
plt.xlabel='Years'
plt.ylabel='Population'
plt.xticks(years)
plt.yticks(pop)
plt.title('Iran Population')
plt.plot(years, pop, ls='--', marker='+', mew=8, lw=2)
```

```python
plt.xlabel='Years'
plt.ylabel='Population'
plt.xticks(years)
plt.yticks(pop, ['21M','28M', '38M', '56M', '66M', '74M', '80M'])
plt.title('Iran Population')
plt.text(1990, 56, "Sixties", fontsize=15)
plt.scatter(years, pop, s=np.array(pop)*5, c=[1, 2, 3, 4, 5, 6, 7])
```

```python
Turkey_pop=[27, 34, 43, 53, 63, 72, 79]

plt.plot(years, pop, ls='--', marker='+', mew=8, lw=2)
plt.plot(years, Turkey_pop, ls=':', marker='o', mew=5, lw=1)
plt.xlabel='Years'
plt.ylabel='Population'
plt.title('Iran and Turkey Populations')
plt.annotate("Sixties", xy=(1990,56), xytext=(1975,56), arrowprops= dict(facecolor='silver', width=4))
plt.legend(['Iran', 'Turkey'], loc='best')
```

```python
plt.subplot(1,2,1)
plt.plot(years, pop, ls='--', marker='+', mew=8, lw=2)
plt.subplot(1,2,2)
plt.plot(years, Turkey_pop, ls=':', marker='o', mew=5, lw=1)
```

```python
def ECDF(data):
  n=len(data)
  x=np.sort(data)
  y=np.arange(1,n+1)/n
  return x, y
```

```python
x, y = ECDF(df['petal.length'])
plt.scatter(x, y)
plt.title('ECDF (petal length)')
plt.margins=0.05
```
